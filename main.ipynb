## Introducción :

El ataque cerebrovascular, también conocido como __ACV__, es una condición médica grave que surge cuando el flujo sanguíneo hacia una parte del cerebro se ve interrumpido, ya sea por un coágulo o la ruptura de un vaso sanguíneo. Este fenómeno puede tener consecuencias devastadoras, afectando funciones vitales como el habla, el movimiento y la cognición. Un mayor entendimiento de esta enfermedad es esencial, ya que la rápida intervención puede marcar la diferencia entre la recuperación y la discapacidad permanente.<br>
Según la Organización Mundial de la Salud (OMS), el accidente cerebrovascular ocupa el preocupante segundo lugar entre las principales causas de mortalidad a nivel global, contribuyendo aproximadamente con el 11% del total de defunciones. Este desafío de salud pública está en constante aumento debido a la adopción generalizada de estilos de vida que descuidan la importancia de hábitos saludables y una alimentación balanceada. En este contexto, la innovación en dispositivos electrónicos que registran los signos vitales de la salud ha allanado el camino hacia el desarrollo de soluciones automatizadas, utilizando técnicas de inteligencia artificial como su núcleo.<br>
Este conjunto de datos se emplea para anticipar la probabilidad de que un paciente experimente un accidente cerebrovascular (ACV), teniendo en cuenta diversos parámetros de entrada como el sexo, la edad, la presencia de diversas enfermedades y el hábito de fumar o tomar alcohol. La capacidad de predecir ACV basándose en los hábitos de las personas no solo permite identificar patrones cruciales para evitar, minimizando así el riesgo de esta patología, sino que también conlleva beneficios tanto a nivel individual como a nivel societal.

El objetivo de este __nootebook__ es identificar patrones de comportamiento claves que predisponen a una persona a experimentar un accidente cerebrovascular (ACV) y desarrollar un modelo que permita predecir la probabilidad individual de sufrir esta condición. Este estudio busca no solo reconocer los factores de riesgo pertinentes, sino también crear una herramienta predictiva que contribuya a una intervención temprana y eficaz, mejorando así las oportunidades de prevención y atención personalizada en el ámbito de la salud cardiovascular.

Como se puede observar hay muchos __outliers__, y como estos datos pueden inferir en mi modelo, decido eliminarlos del __data frame__ original y guardarlos en otro __data frame__ para su posterior análisis. <br>
Para la eliminación uso el método del __Criterio del rango intercuartílico (IQR)__.

```python
# Esta función me retorna los indices de los registros que son considerados outliers debido al criterio del rango intercuartilico
def outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 * IQR)

    ls = df.index[(df[col] < lower_bound) | (df[col] > upper_bound)]

    return ls
```

```python
# Esta función recibe un DataFrame y una lista de indices, y elimina los registros de la lista de indices
def remove(df, ls):
    ls = sorted(set(ls))
    df = df.drop(ls)

    return df
```

```python
# Este bloque de codigo genera una lista de indices (index_list) con los indices considerados outliers por el metodo IQR, decidí utilizar un set() para no tener una lista con valores repetidos
indexs = set()

for col in number_columns:
    indexs.update(set(outliers(df, col)))

index_list = list(indexs)
```

```python
# Guardo los outliers en otro DataFrame
df_outliers = df.iloc[sorted(index_list)].reset_index(drop=True)
df_outliers
```

```python
df_outliers.shape # 13651 registros son considerados outliers
```

```python
# Elimino los outliers del dataset original 
df = remove(df, index_list).reset_index(drop=True)
df
```

Realizo nuevamente el gráfico de los boxplots para ver los cambios despues de haber eliminado los outliers.

```python
# Libreria utilizada
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
```

En esta sección codifico las variables categoricas, a todas las variables con dos categorias las codifique con __Label Encoder__ para una mejor manipulación. Mientras que a las demas variables las codifique con el metodo __One Hot Encoding__ ya son variables categoricas nominales, excepto a la variable __AgeCategory__ que la codifique con el metodo __Ordinal Encoder__ ya que es una variable categorica ordinal.

```python
# DataFrame que contiene las variables categoricas con dos categorias
df_two_categ = df[two_categories + ['CovidPos']]

# Codifico el DataFrame
# Sex = {0: Female, 1:Male}
# Todos los demas casos {0: No, 1:Yes}
warnings.filterwarnings('ignore')
le = LabelEncoder()
for col in df_two_categ.columns:
    df_two_categ[col] = le.fit_transform(df_two_categ[col])

df_two_categ
```

```python
# Codifico las variables categoricas con más de 2 categorias excepto la variable AgeCategory
df_one_hot = pd.get_dummies(df[many_categories[1:-1]]).astype(int)
df_one_hot
```

```python
#Codifico la varaible AgeCategory
# Orden de codificación
order = df['AgeCategory'].value_counts().sort_index().index.tolist()

# Defino el DataFrame
df_age = df[['AgeCategory']]

# Ordinal Encoder
ordinal_encoder = OrdinalEncoder(categories=[order])

df_age['AgeCategory'] = ordinal_encoder.fit_transform(df[['AgeCategory']]).astype(int)
df_age
```

```python
# Armo mi DataFrame final
df_stroke = pd.concat([df[number_columns], df_two_categ, df_one_hot, df_age], axis=1)
df_stroke
```
